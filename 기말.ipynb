31. [코딩 문제] 주어진 데이터를 이용하여 로지스틱 분류 회귀(Logistic regression) 모델링을 코딩하시오. 아래 문제별 코드를 실행하면 로지스틱 회귀 분류 모델링의 각 단계를 확인할 수 있도록 코딩하시오.

```python
# (31-1) 주어진 데이터를 사용하여 로지스틱 회귀 모델을 생성하고 지도학습으로 훈련하시오.

import numpy as np
from sklearn.linear_model import LogisticRegression

# 주어진 데이터
X_train = np.array([[1], [2], [3], [4], [5]])
y_train = np.array([0, 0, 1, 1, 1])

# 모델 생성 및 훈련
model = LogisticRegression()
model.fit(X_train, y_train)

# 모델 평가
print("훈련 세트 점수: {:.2f}".format(model.score(X_train, y_train)))
```

```python
# (31-2) 문제 (31-1)에서 얻은 로지스틱 분류 모델의 회귀 계수와 y절편을 출력하시오.

# 회귀 계수와 y절편 출력
print("회귀 계수: ", model.coef_)
print("y절편: ", model.intercept_)
```

```python
# (31-3) 주어진 특성 데이터를 사용하여 문제 (31-1)에서 얻은 로지스틱 회귀 모델의 예측 값
# 을 추정하시오.

# 주어진 특성 데이터
X_test = np.array([[6], [7]])

# 모델을 사용한 예측
predictions = model.predict(X_test)

# 예측 결과 출력
print("예측 값: ", predictions)
```

```python
# (31-4) 주어진 특성 데이터를 사용하여 문제 (31-1)에서 얻은 로지스틱 회귀 모델의 분류 성
# 능 평가를 confusion matrix와 classification_report을 활용하여 출력하시오.

from sklearn.metrics import confusion_matrix, classification_report

# 실제 값
y_test = np.array([0, 1])  # 이 값은 실제 테스트 데이터의 레이블이어야 합니다.

# 모델을 사용한 예측
predictions = model.predict(X_test)

# confusion matrix 출력
print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))

# classification report 출력
print("\nClassification Report:")
print(classification_report(y_test, predictions))
```

32. 영화 평점 이진분류 모델의 테스트 데이터 셋에 대한 혼동 행렬(confusion matrix)이 다
음과 같다. 혼동 행렬의 숫자( 테스트 데이터 셋의 샘플 개수)는 다음과 같이 정의한다.
∙TP (True Positive): 2 (실제 good이며 예측도 good)
∙FN (False Negative): 1 (실제 good이지만 예측은 bad)
∙FP (False Positive): 1 (실제 bad이지만 예측은 good)
∙TN (True Negative): 1 (실제 bad이며 예측도 bad)


(32-1) 혼동 행렬을 사용하여 정확도(accuracy), 정밀도(Precision), 재현율(Recall)를 계산하시오

= 정확도 : 모델이 얼마나 정확하게 분류했는지 나타냄 
Accuracy = (TP+TN) / (TP+TN+FP+FN) = 3/5
= 정밀도 : 실제로 양성으로 예측된 샘플 중에서 실제 양성의 비율 :
Precision = TP / (TP+FP) = 2/3
= 재현율은 실제 양성 샘플 중에서 양성으로 예측된 샘플의 비율
답: Recall= TP /(TP+FN) = 2/3
